```python
from ultralytics import YOLO

# Load YOLOv12 model
model = YOLO('yolo12n.pt')   # change to s/n/l as needed

# Train the model
results = model.train(
    data="/content/drive/MyDrive/CarDD/data.yaml",
    epochs=30,
    imgsz=640,
    batch=16,          # SAFE VALUE for T4
    workers=1,
    # Saving
    save=True,
    save_period=-1,
    project="/content/drive/MyDrive/yolo_training",
    name="yolov12_custom_cardd",
    exist_ok=True,
)

```
